```python
#### go for the out put
import pandas as pd
import csv

# This function return 5 most similar data to a particular passed input 
# Euclidean distance
def best_five(difference,ifile_1):
    
    courses=[]
    contact_details=[]
    i=0
    while(len(courses)<5 and i<len(difference)):
        ifile_1.seek(0)
        reader_alumini=csv.reader(ifile_1)
        next(reader_alumini)
        for row_alumini in reader_alumini:
            if difference[i][0]==row_alumini[3] and row_alumini[64] not in courses:
                
                courses.append(row_alumini[64])
                contact_details.append(list(row_alumini[0:1]+row_alumini[12:13]+row_alumini[15:16]))
        i+=1
    
    return courses,contact_details

def myfunction(row_student,ifile_1):
    ifile_1.seek(0)
    reader_alumini=csv.reader(ifile_1)
    next(reader_alumini)
    
    
    difference=[]
    for row_alumini in reader_alumini:

        diff=0
        if row_alumini[4]==row_student[4]:

            diff+=(float(row_alumini[18])-float(row_student[18]))**2
            for i in range(35,55):
                diff+= (ord(row_alumini[i])-ord(row_student[i]))**2
            diff/=float(row_alumini[65])
#           
            empty=[]
            empty.append(row_alumini[3])
            empty.append(diff)
            difference.append(empty)
            
    difference.sort(key=lambda x: x[1])

    
    return best_five(difference,ifile_1)
    
def recommender_function(alumini,data):          
    ifile_1=open(alumini,'r',newline='')
    ifile_2=open(data,'r',newline='')
    #output file for recommendation
    ofile_2=open('recommeded_courses.csv','w',newline='')
    #output file foralumini details
    ofile_3=open('alumini_contact_file.csv','w',newline='')
    reader_student=csv.reader(ifile_2)
    
    writer=csv.writer(ofile_2)
    writer_alumini_contact=csv.writer(ofile_3)
#     for course recommedation
    row=['name','student_no','branch','recommed_course_1','recommed_course_2','recommed_course_3','recommed_course_4','recommed_course_5']
    writer.writerow(row)
    next(reader_student)
    
#     for alumini contact
    row_alumini_contact=['student_no','al_name_1','al_contact_1','al_email_1','al_name_2','al_contact_2','al_email_2','al_name_3','al_contact_3','al_email_3','al_name_4','al_contact_4','al_email_4','al_name_5','al_contact_5','al_email_5']
    writer_alumini_contact.writerow(row_alumini_contact)
    for row in reader_student:
        recommender=[]
#         name
        recommender.append(row[0])
#         student_no
        recommender.append(int(float(row[3])))
#         branch
        recommender.append(row[4])
        (re_courses,alumini_contact)=myfunction(row,ifile_1)
#         inserting courses
        recommender+=re_courses
        writer.writerow(recommender)
        
#         inserting contact details        
        flattened_list = [y for x in alumini_contact for y in x]
#         print(flattened_list)
        flattened_list.insert(0, int(row[3]))
        writer_alumini_contact.writerow(flattened_list)
        
        
    ifile_1.close()
    ifile_2.close()
    ofile_2.close()
    ofile_3.close()
    
    ifile=pd.read_csv('recommeded_courses.csv')
    
    return ifile
```
