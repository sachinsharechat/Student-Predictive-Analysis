```python
import nbimporter
# from question_feedback import add_to_head
import csv
import random

def a_func(num):
    if num==1:
        return 'YES'
    else:
        return 'NO'
def count_magic(row,i,j,boolean):
    count=0
    if boolean:
        while i<=j:
            if row[i]=='C':
                count+=1
            i+=1
    else:
        while i<=j:
            if row[i]=='A' :
                count+=1
            i+=1   
    return count
        
def assigning_label(row):
    row_complete=row
    row_complete.append('0')
   
    drop=count_magic(row,35,44,False)
    teacher=count_magic(row,54,62,True)
#     assigning
    if drop+teacher>=9 and float(row[18])<=60:
        row_complete[-1]= a_func(random.randint(0,1))
    else:
        row_complete[-1]='NO'
    return row_complete
        
    
ifile=open('clean_data/clean_4.csv','r',newline='')
ofile=open('dropout_data_4.csv','w',newline='')
reader = csv.reader(ifile)
writer = csv.writer(ofile)
row_head=next(reader)
# print(row_head)
# making the header of csv
row_head.append('dropout')
# writing the header
writer.writerow(row_head)
# main course
for row in reader:
    row_complete=assigning_label(row)
    writer.writerow(row_complete)
  
    

ifile.close()
ofile.close()
```
