```python
import csv
import random

def add_to_head(row_head):
#     adding dropout question
    for i in range(1,11):
        row_head.append('ques_S_drop_'+str(i))
#     adding personality qeustion
    for i in range(1,11):
        row_head.append('ques_S_per'+str(i))
#     adding teacher feedback
    for i in range(1,9):
        row_head.append('ques_T_feed_'+str(i))
def choose(i,j):
    num=random.randint(i,j)
    if num==1:
        return 'A'
    elif num==2:
        return 'B'
    elif num==3:
        return 'C'
    elif num==4:
        return 'D'
    
def myfunction(row_complete,test_type,st_type,i,j):
    while i<=j:
        if test_type=='D': 
            if st_type=='bad':
                row_complete[i]=choose(1,2)
            else:
                row_complete[i]=choose(1,4)
        elif test_type=='P': 
            if st_type=='bad':
                row_complete[i]=choose(1,3)
            else:
                row_complete[i]=choose(1,3)
        elif test_type=='T': 
            if st_type=='bad':
                row_complete[i]=choose(2,3)
            else:
                row_complete[i]=choose(1,2)
        
        i+=1
        
        
def assigning_values(row):
    row_complete=row
    
    dummy=[0]*28
    row_complete+=dummy
#     filling for dropout
    if float(row[18])<65:
        row_complete[35]=choose(1,1)
        row_complete[44]=choose(1,1)
        myfunction(row_complete,'D','bad',36,43)
        myfunction(row_complete,'P','bad',45,54)
        myfunction(row_complete,'T','bad',55,62)
    else:
        row_complete[35]=choose(1,2)
        row_complete[44]=choose(1,2)
        myfunction(row_complete,'D','good',36,43)
        myfunction(row_complete,'P','good',45,54)
        myfunction(row_complete,'T','good',55,62)
    return row_complete
    

ifile=open('student_data/student_data_EC_2.csv','r',newline='')
ofile=open('final_data_2.csv','a',newline='')
reader = csv.reader(ifile)
writer = csv.writer(ofile)
# reading the first row from csv
row_head=next(reader)

# making the header of csv
#add_to_head(row_head)
# writing the header
#writer.writerow(row_head)

# main course
for row in reader:
    if len(row)==35:
        row_complete=assigning_values(row)
        writer.writerow(row_complete)

ifile.close()
ofile.close()
```
