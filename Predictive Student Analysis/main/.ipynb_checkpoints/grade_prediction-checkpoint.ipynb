```python
import pandas as pd
import numpy as np
file=pd.read_csv('alumini_data.csv')
# file.columns
```

```python
work=pd.get_dummies(file,columns=['branch','category', 'hostler', 'Addmission_mod','ques_S_drop_1', 'ques_S_drop_2', 'ques_S_drop_3',
       'ques_S_drop_4', 'ques_S_drop_5', 'ques_S_drop_6', 'ques_S_drop_7',
       'ques_S_drop_8', 'ques_S_drop_9', 'ques_S_drop_10','ques_T_feed_1', 'ques_T_feed_2', 'ques_T_feed_3',
       'ques_T_feed_4', 'ques_T_feed_5', 'ques_T_feed_6', 'ques_T_feed_7',
       'ques_T_feed_8'],drop_first=True)
# work.to_csv('numeric.csv', sep=',',index=False)
# for i in work.columns:
#     print("'"+i+"',")
```

```python
from sklearn.cross_validation import train_test_split
X=work.loc[:, ['high_school',
'intermediate',
'B_tech',
'sem_1_marks',
'sem_1_attendance',
'sem_2_marks',
'sem_2_attendance',
'sem_3_attendance',
'branch_ECE',
'branch_IT',
'branch_ME',
'category_OBC',
'category_SC',
'hostler_YES',
'Addmission_mod_UPSEE','ques_S_drop_1_B',
'ques_S_drop_2_B',
'ques_S_drop_2_C',
'ques_S_drop_2_D',
'ques_S_drop_3_B',
'ques_S_drop_3_C',
'ques_S_drop_3_D',
'ques_S_drop_4_B',
'ques_S_drop_4_C',
'ques_S_drop_4_D',
'ques_S_drop_5_B',
'ques_S_drop_5_C',
'ques_S_drop_5_D',
'ques_S_drop_6_B',
'ques_S_drop_6_C',
'ques_S_drop_6_D',
'ques_S_drop_7_B',
'ques_S_drop_7_C',
'ques_S_drop_7_D',
'ques_S_drop_8_B',
'ques_S_drop_8_C',
'ques_S_drop_8_D',
'ques_S_drop_9_B',
'ques_S_drop_9_C',
'ques_S_drop_9_D',
'ques_S_drop_10_B',
'ques_T_feed_1_B',
'ques_T_feed_1_C',
'ques_T_feed_2_B',
'ques_T_feed_2_C',
'ques_T_feed_3_B',
'ques_T_feed_3_C',
'ques_T_feed_4_B',
'ques_T_feed_4_C',
'ques_T_feed_5_B',
'ques_T_feed_5_C',
'ques_T_feed_6_B',
'ques_T_feed_6_C',
'ques_T_feed_7_B',
'ques_T_feed_7_C',
'ques_T_feed_8_B',
'ques_T_feed_8_C']]
X=X.dropna(axis=0, how='any')
y=work.dropna(axis=0,subset=['high_school',
'intermediate',
'B_tech',
'sem_1_marks',
'sem_1_attendance',
'sem_2_marks',
'sem_2_attendance',
'sem_3_attendance',

'category_OBC',
'category_SC',
'hostler_YES',
'Addmission_mod_UPSEE']).sem_3_marks
y.dropna(axis=0,how='any',inplace=True)
X.shape
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
# print(X.shape,y.shape)
X.dtypes
```

```python
from sklearn.linear_model import LinearRegression
liner=LinearRegression()
liner.fit(X,y)
```

```python
y_predict=liner.predict(X)
for i,j in zip(y,y_predict):
    print(i,' ',j)
print(y.shape,y_predict.shape)
```

```python

from sklearn import metrics
np.sqrt(metrics.mean_squared_error(y,y_predict))
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.plot(y, liner.predict(X),color='k')

plt.show()
```

```python
import pickle
pickle.dump(liner, open('grade_prediction_model', 'wb'))
```
