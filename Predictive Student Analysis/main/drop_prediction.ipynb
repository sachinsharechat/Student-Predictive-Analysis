```python
import pandas as pd
import numpy as np
file=pd.read_csv('alumini_data.csv')
file.columns
```

```python
X=file.loc[:,['branch','category', 'hostler', 'Addmission_mod',
       'B_tech', 'sem_1_marks', 'sem_1_attendance', 'sem_2_marks',
       'sem_2_attendance', 'sem_3_attendance','ques_S_drop_1', 'ques_S_drop_2', 'ques_S_drop_3',
       'ques_S_drop_4', 'ques_S_drop_5', 'ques_S_drop_6', 'ques_S_drop_7',
       'ques_S_drop_8', 'ques_S_drop_9', 'ques_S_drop_10', 'ques_S_per1',
       'ques_S_per2', 'ques_S_per3', 'ques_S_per4', 'ques_S_per5',
       'ques_S_per6', 'ques_S_per7', 'ques_S_per8', 'ques_S_per9',
       'ques_S_per10', 'ques_T_feed_1', 'ques_T_feed_2', 'ques_T_feed_3',
       'ques_T_feed_4', 'ques_T_feed_5', 'ques_T_feed_6', 'ques_T_feed_7',
       'ques_T_feed_8','dropout']]

X=X.dropna(axis=0, how='any')
y=X.dropout

X.drop(['dropout'],axis=1,inplace=True)

X=pd.get_dummies(X,columns=['branch','category', 'hostler', 'Addmission_mod','ques_S_drop_1', 'ques_S_drop_2', 'ques_S_drop_3',
       'ques_S_drop_4', 'ques_S_drop_5', 'ques_S_drop_6', 'ques_S_drop_7',
       'ques_S_drop_8', 'ques_S_drop_9', 'ques_S_drop_10', 'ques_S_per1',
       'ques_S_per2', 'ques_S_per3', 'ques_S_per4', 'ques_S_per5',
       'ques_S_per6', 'ques_S_per7', 'ques_S_per8', 'ques_S_per9',
       'ques_S_per10', 'ques_T_feed_1', 'ques_T_feed_2', 'ques_T_feed_3',
       'ques_T_feed_4', 'ques_T_feed_5', 'ques_T_feed_6', 'ques_T_feed_7',
       'ques_T_feed_8'],drop_first=True)
y1=file.loc[:,['dropout']]
y=pd.get_dummies(y,columns=['dropout'],drop_first=True)

print(X.shape)
print(y.shape)

# X.to_csv('dummy.csv', sep=',',index=False)
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)

log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_predict=rf.predict(X)
from sklearn import metrics
np.sqrt(metrics.mean_squared_error(y,y_predict))
for i,j in zip(y_predict,y.YES):
    print(i,' ',j)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.scatter(y, log.predict(X),color='k')

plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y, y_predict))
```

```python
import pickle
pickle.dump(log,open('dropout_prediction_model','wb'))
pickle.dump(rf, open('dropout_prediction_model_random_forest', 'wb'))
```
