```python
import csv
import random
def course_selector(branch,i,j):
    course_dict={'CSE':['Mtech in Security and Forensic', 'Mtech in Image Processing', 'Startup', 'Mtech in Data Analytics', 'Mtech in Artificial Intelligence', 'MS in Algorithms', 'JOB in Software Industry', 'MBA', 'Govt. Job and PSU', 'MS in Computer Architecture', 'Post Grad in Computer Networking', 'Mtech in Cloud Computing', 'Post Grad in Machine Learning and Statistics', 'MS in Applied Science and Mathematics', 'MS in Environment and Ecology', 'Post Grad in Robotics'],'IT':['Mtech in Cloud Computing', 'Post Grad in Machine Learning and Statistics', 'MSc in Applied Science and Mathematics', 'MS in Environment and Ecology', 'Post Grad in Robotics', 'JOB in Software Industry', 'MBA', 'Govt. Job and PSU', 'MS in Computer Architecture', 'Post Grad in Computer Networking', 'Mtech in Electronics Circuit Design', 'MSc in Distributed System', 'MS in Parallel Processing', 'Mtech in Artificial Intelligence', 'Mtech in Space Science', 'MS in Cosmology'],'ME':['Mtech in Design engineering', 'Govt. Job and PSU', 'Mtech in fluidics/ Hydraulics', 'Mtech in Manufacturing engineering', 'MBA', 'JOB in Mechanical Industry', 'Startup', 'MSc in Mineral Science', 'JOB in Software Industry', 'Mtech in Automobile Engineering', 'MS in Acoustical Engineering', 'MS in Aerospace Engineering', 'Post Grad in Marine Engineering', 'Master in Mechatronics', 'Post Grad in Thermal Engineering', 'Mtech in Energy Engineering'],'ECE':['Mtech in Electronics Circuit Design', 'MSc in Signal processing', 'Mtech in Robotics', 'MSc in Embedded Systems', 'MSc Digital Systems', 'JOB in Software Industry', 'Govt. Job and PSU', 'MBA', 'Startup', 'MS in Electronics and Communication', 'MS in Applied Mathematics', 'Post Grad in Automative Engineering', 'Post Grad in Electrical Power Engineering', 'MS in Telecommunication', 'MSc in Statistics']}
    num=random.randint(i,j)
    return course_dict[branch][num]

def myfunction(row):
    complete_row=row
    complete_row+=[0,0]
#     print(complete_row)
    if row[63]=='YES':
        num=random.randint(1,5)
        if num==1:
            pg_course='Startup'
            rating=random.randint(1,10)
        elif num==2:
            pg_course='Job'
            rating=random.randint(1,8)
        elif num==3:
            pg_course='BBA'
            rating=random.randint(1,8)    
        elif num==4:
            pg_course='BSc'
            rating=random.randint(1,8)    
        else:
            pg_course='Different Undergraduate Course'
            rating=random.randint(1,8)
    else:
        if float(row[18])>=85:
            pg_course=course_selector(row[4],0,7)
            rating=random.randint(5,10)
        elif float(row[18])>=65: 
            pg_course=course_selector(row[4],0,8)
            rating=random.randint(3,10)
        else:
            pg_course=course_selector(row[4],2,7)
            rating=random.randint(1,8)
        
    complete_row[64]=pg_course
    complete_row[65]=rating
    return list(complete_row)     
            
        
        
    
ifile=open('dropout_data_4.csv','r',newline='')
ofile=open('alumini_data.csv','w',newline='')
reader=csv.reader(ifile)
writer=csv.writer(ofile)
head_row=next(reader)
head_row.append('after_grad')
head_row.append('rating')
writer.writerow(head_row)

for row in reader:
    complete_row=myfunction(row)
    writer.writerow(complete_row)
    

ifile.close()
ofile.close()
```
